Solving XOR problem using deep feed forward network

```python
# demonstrates how to solve the XOR problem using a deep feedforward neural network, which is constructed from multiple perceptrons.
# The XOR problem is a classic example in machine learning because a single perceptron cannot solve it due to its non-linear separability.
# A multi-layer perceptron (like the one built here with AND, OR, and NOT gates) is required.
```

```python
# Importing the NumPy library, which is essential for numerical operations, especially for array manipulations and dot products in our perceptron model.
import numpy as np
```

```python
# This function defines a basic Unit Step activation function.
# It outputs 1 if the input `v` is greater than or equal to 0 (neuron fires).
# It outputs 0 if the input `v` is less than 0 (neuron does not fire).
def unitStep(v):
    if v >= 0:
        return 1
    else:
        return 0
```

```python
# This function defines a simple Perceptron model.
# A perceptron takes inputs (x), multiplies them by weights (w), adds a bias (b),
# and then passes the result through an activation function (unitStep).
# `x` is the input vector.
# `w` is the weight vector.
# `b` is the bias term.
def perceptronModel(x, w, b):
    # Calculate the weighted sum of inputs plus bias (v = w * x + b)
    v = np.dot(w, x) + b
    # Apply the unit step activation function to get the output (y)
    y = unitStep(v)
    return y
```

```python
# This function implements the NOT logic gate using a perceptron.
# A NOT gate has one input and outputs the inverse of that input.
# For example, NOT(0) = 1, NOT(1) = 0.

# The weight (wNOT) is -1 and the bias (bNOT) is 0.5.
# If x = 0: v = (-1 * 0) + 0.5 = 0.5 -> unitStep(0.5) = 1 (Correct: NOT 0 is 1)
# If x = 1: v = (-1 * 1) + 0.5 = -0.5 -> unitStep(-0.5) = 0 (Correct: NOT 1 is 0)
def NOT_logicFunction(x):
    wNOT = -1
    bNOT = 0.5
    return perceptronModel(x, wNOT, bNOT)
```

```python
# This function implements the AND logic gate using a perceptron.
# An AND gate has two inputs and outputs 1 only if both inputs are 1.
# Otherwise, it outputs 0.

# The weights (w) are [1, 1] and the bias (bAND) is -1.5.
# Example for input [0, 0]: v = (1*0 + 1*0) + (-1.5) = -1.5 -> unitStep(-1.5) = 0 (Correct)
# Example for input [0, 1]: v = (1*0 + 1*1) + (-1.5) = -0.5 -> unitStep(-0.5) = 0 (Correct)
# Example for input [1, 0]: v = (1*1 + 1*0) + (-1.5) = -0.5 -> unitStep(-0.5) = 0 (Correct)
# Example for input [1, 1]: v = (1*1 + 1*1) + (-1.5) = 0.5 -> unitStep(0.5) = 1 (Correct)
def AND_logicFunction(x):
    w = np.array([1, 1])
    bAND = -1.5
    return perceptronModel(x, w, bAND)
```

```python
# This function implements the OR logic gate using a perceptron.
# An OR gate has two inputs and outputs 1 if at least one input is 1.
# It outputs 0 only if both inputs are 0.

# The weights (w) are [1, 1] and the bias (bOR) is -0.5.
# Example for input [0, 0]: v = (1*0 + 1*0) + (-0.5) = -0.5 -> unitStep(-0.5) = 0 (Correct)
# Example for input [0, 1]: v = (1*0 + 1*1) + (-0.5) = 0.5 -> unitStep(0.5) = 1 (Correct)
# Example for input [1, 0]: v = (1*1 + 1*0) + (-0.5) = 0.5 -> unitStep(0.5) = 1 (Correct)
# Example for input [1, 1]: v = (1*1 + 1*1) + (-0.5) = 1.5 -> unitStep(1.5) = 1 (Correct)
def OR_logicFunction(x):
    w = np.array([1, 1])
    bOR = -0.5
    return perceptronModel(x, w, bOR)
```

```python
# This function solves the XOR (Exclusive OR) problem by combining the AND, OR, and NOT logic gates.
# A single perceptron cannot solve XOR because it's not linearly separable.
# The XOR logic can be expressed as: A XOR B = (A AND NOT B) OR (NOT A AND B)
# Or, more simply, A XOR B = (A OR B) AND (NOT (A AND B))

# The implementation here uses the latter approach:
# 1. y1 = A AND B
# 2. y2 = A OR B
# 3. y3 = NOT (y1)  (which is NOT (A AND B))
# 4. finalOutput = y2 AND y3 (which is (A OR B) AND (NOT (A AND B)))
def XOR_logicFunction(x):
  # Calculate A AND B
  y1 = AND_logicFunction(x)
  # Calculate A OR B
  y2 = OR_logicFunction(x)
  # Calculate NOT (A AND B)
  y3 = NOT_logicFunction(y1)
  # Create a new input vector for the final AND gate: [ (A OR B), (NOT (A AND B)) ]
  final_x = np.array([y2, y3])
  # Calculate (A OR B) AND (NOT (A AND B)) to get the XOR result
  finalOutput = AND_logicFunction(final_x)
  return finalOutput
```

```python
# This section tests the implemented XOR_logicFunction with all possible binary inputs for two variables.

# Define the four possible input combinations for two binary variables (0 or 1).
test1 = np.array([0, 1])
test2 = np.array([1, 1])
test3 = np.array([0, 0])
test4 = np.array([1, 0])

# Print the results of the XOR function for each test case.
# Expected XOR outputs:
# XOR(0, 1) = 1
# XOR(1, 1) = 0
# XOR(0, 0) = 0
# XOR(1, 0) = 1
print("XOR({}, {}) = {}".format(0, 1, XOR_logicFunction(test1)))
print("XOR({}, {}) = {}".format(1, 1, XOR_logicFunction(test2)))
print("XOR({}, {}) = {}".format(0, 0, XOR_logicFunction(test3)))
print("XOR({}, {}) = {}".format(1, 0, XOR_logicFunction(test4)))
```
